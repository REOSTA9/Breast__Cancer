```python
import pandas as pd
from apriori import apriori
from transactionencoder import TransactionEncoder
from association_rules import association_rules
```

```python
## LOAD DATA ##
```

```python
"""Compute candidate 1-itemset"""
C1 = {}
"""total number of transactions contained in the file"""
transactions = 0
D = []
T = []
with open("data.csv", "r") as f:
    for line in f:
        T = []
        transactions += 1
        for word in line.split():
            T.append(word)
            if word not in C1.keys():
                C1[word] = 1
            else:
                count = C1[word]
                C1[word] = count + 1
        D.append(T)
```

```python
## TRANSFORM DATA ##
```

```python
te = TransactionEncoder()
te_ary = te.fit(D).transform(D)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

```python
## COMPUTE FREQUENT ITEM ##
```

```python
frequent_itemsets = apriori(df, min_support=0.05, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets.head()
```

```python
rules = association_rules(frequent_itemsets)
rules
```

```python
## ADD ANTECEDENTS LENGTH ##
```

```python
rules["antecedant_len"] = rules["antecedents"].apply(lambda x: len(x))
rules.head()
```
